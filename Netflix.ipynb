# Netflix Data Cleaning

```python
import pandas as pd
```

```python
df = pd.read_csv('DataSets/netflix.csv')
```

```python
df
```

```python
df.info()
```

```python
"""
Delecting redundant columns
Renaming the columns
Dropping duplicate rows
Remove the NaN values in the dataset
Check the some more Transformations
"""
```

```python
df.drop(columns='rating', inplace=True)
```

```python
df.columns = df.columns.str.capitalize()
```

```python
df.head()
```

# Drop Duplicates

```python
df.duplicated().sum()
```

```python
# for cleaning duplicate
df.drop_duplicates(inplace=True)
```

# Remove the NaN values from the datasets

```python
df.isna().sum()
```

```python
# If we had na values
df.dropna(inplace=True)
```

```python
# df['Show_id'] = df['Show_id'].apply(lambda x: x.split('s')[1])
df['Show_id'] = df["Show_id"].str.replace('s', '', regex=False).astype(int)
```

```python
df['Show_id']
```

```python
df['Date_added'] = df['Date_added'].str.replace('/', '-')
# df["Date_added"] = df["Date_added"].apply(lambda x: x.replace("/","-"))
```
