# Deleting Redundent Columns

```python
df.columns
```

```python
columns_to_keep = ['name', 'online_order', 'book_table', 'rate','dish_liked', 'approx_cost(for two people)']
columns_to_drop = ['url', 'address', 'votes', 'phone', 'location','cuisines', 'reviews_list', 'menu_item', 'listed_in(type)', 'listed_in(city)']
```

```python
df.columns
```

```python
len(columns_to_keep), len(columns_to_drop)
```

# 1st Way filtering Data

```python
modify_df = df[columns_to_keep]
modify_df
```

# 2nd way, droping the columns

```python
df.drop(columns=columns_to_drop, inplace=True)
```

```python
df
```

# Rename The columns

```python
df.columns = df.columns.str.upper().str.replace(' ', '_')
```

```python
df.columns
```

# Dropping Duplicates

```python
df.duplicated().sum()
```

```python
df.duplicated().value_counts()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

# Remove the NaN Value from DataSets

```python
df.isna().sum()
```

```python
df.drop(columns=['REST_TYPE'], inplace=True)
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.reset_index(inplace=True)
df.drop(columns=['index'], inplace=True)
df.drop(columns=['level_0'], inplace=True)
```

# Cleaning Indivitual columns

```python
df["RATE"]
```

```python
df["RATE"] = df["RATE"].str.replace(' ','').str.strip()
```

```python
df["RATE"] = df["RATE"].str.replace('/',' out of ')
```

```python
df
```

```python
df['ONLINE_ORDER'] = df['ONLINE_ORDER'].map({'Yes':1, 'No':0,})
```
